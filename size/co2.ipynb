```python
import numpy as np
from matplotlib import pyplot as plt
def get_data(data):
    n = []
    e = []
    err = []
    for line in data.splitlines():
        if len(line) > 0:
            m = [int(line.split()[0])]
            n = np.hstack((n,m))
            energy = [float(line.split()[-3])]
            e = np.hstack((e,energy))
            error = [float(line.split()[-1])]
            err = np.hstack((err,error))
    e_per = e/n
    err_per = err/n
    return n,e_per,err_per
```

```python
# CO2 sto6g
data_cisd = '''
1 AFQMC energy: -187.062842 +/- 0.000217
2 AFQMC energy: -374.125756 +/- 0.000337
4 AFQMC energy: -748.253198 +/- 0.000689
8 AFQMC energy: -1496.513537 +/- 0.002032
'''
data_rhf = '''
1  AFQMC energy: -187.073167 +/- 0.000554
2  AFQMC energy: -374.145279 +/- 0.000787
4  AFQMC energy: -748.289147 +/- 0.001446
8  AFQMC energy: -1496.577854 +/- 0.002263
'''
data_pt = '''
1 AFQMC energy: -187.056936 +/- 0.000312
2 AFQMC energy: -374.113213 +/- 0.000547
4 AFQMC energy: -748.227116 +/- 0.001017
8 AFQMC energy: -1496.453034 +/- 0.002224
'''
data_pt2 = '''
1 AFQMC energy: -187.055907 +/- 0.000409
2 AFQMC energy: -374.110792 +/- 0.000658
4 AFQMC energy: -748.221633 +/- 0.001143
8 AFQMC energy: -1496.446834 +/- 0.003274
'''
n1,e_per1,err_per1 = get_data(data_rhf)
n2,e_per2,err_per2 = get_data(data_cisd)
n3,e_per3,err_per3 = get_data(data_pt)
n4,e_per4,err_per4 = get_data(data_pt2)
# plt.errorbar(n1,e_per1,yerr=err_per1,fmt='.',capsize=5,label='AFQMC/HF')
# plt.plot(n1,e_per1, '--', color='C0', alpha=0.3)
plt.errorbar(n2,e_per2,yerr=err_per2,fmt='.',color='C1',capsize=5,label='AFQMC/CISD')
plt.plot(n2,e_per2, '--', color='C1', alpha=0.3)
plt.errorbar(n3,e_per3,yerr=err_per3,fmt='.',color='C2',capsize=5,label='AFQMC/CCSD_PT')
plt.plot(n3,e_per3, '--', color='C2', alpha=0.3)
plt.errorbar(n4,e_per4,yerr=err_per4,fmt='.',color='C3',capsize=5,label='AFQMC/CCSD_PT2')
plt.plot(n4,e_per4, '--', color='C3', alpha=0.3)
plt.xlabel('number of H2O dimer')
plt.ylabel('energy per dimer')
plt.title('AFQMC Size-Extensivity')
plt.legend()
plt.show()
```
